```python
# -- Compute Indicator --

#data = indic.get_historic_data("ETH-EUR", period="7d", interval="15m")

# -- Get Data --
# ticker = yf.Ticker("^FCHI") #ETH - BTC #XLM
ticker = yf.Ticker("ETH-USD") #ETH - BTC #XLM

def compute_things(data):
    #data = ticker.history(start="2020-12-19", end="2020-12-21", interval="1m")

    data['tema'] = indic.SMA(data['Close'], span = 200)
    data['sma'] = indic.EMA(data['Close'], span = 50)
    data['ss'] = indic.EMA(data['Close'], span = 1)

    data['buy_line'] = indic.EMA(indic.MACD(data['Close']), span = 10)# 10 10 9
    data['sell_line'] = indic.MACD(data['Close'])


    data['mom2'] = indic.momentum2(data['tema'], span = 1)
    data['mom'] = indic.momentum2(data['sell_line'], span = 1)
    data['mom_buy'] = indic.momentum2(data['buy_line'], span = 1)

    data['flat'] = (data['sell_line']-data['buy_line'])
    data['test'] = data['flat'].rolling(window = 5).sum()

    data.dropna(inplace=True)
    return data
```

```python
def compute(buy, sell, money, fees, output=False, details=False):
    """ buy (list), sell (list) """ 
    nb_coin = 0
    start_money = money
    money = money
    for i in range(len(buy)):
        nb_coin, money = do_buy(buy['Close'][i], money, fees)
        #if details: print(">> buy at:  " + str(buy['Close'][i]))
        coin_buyed = nb_coin
        nb_coin, money = do_sell(sell['Close'][i], nb_coin, fees)
        #if details: print(">> sell at: " + str(sell['Close'][i]))

        diff = sell['Close'][i] - buy['Close'][i]
        b = buy['Close'][i]
        s = sell['Close'][i]
        d = indic.print_pct(indic.pct_change(buy['Close'][i], sell['Close'][i]))
        if details: print(f'  ▷ Buy {coin_buyed:.2f} at: {b:.2f} | Sell at: {s:.2f} >>> {d}')
        #if details: print(f'▶ Buy: {b:.2f} -- {s:.2f} :Sell || {d}') 

    diff = money-start_money
    if output:
        return money, diff
    else:
        print(f'▶ Start with: {start_money} | End with: {money} >>> Gain: {diff:.2f} - {indic.print_pct(diff)}')
#compute(buy,sell)   

def do_buy(price, money, fees):
    """ substract the fees to the amount of money and compute how many coin you can get """
    nb_coin = (money - indic.pct_of(money, fees)) / price
    money = 0
    return nb_coin, money
    
def do_sell(price, nb_coin, fees):
    money = (price * nb_coin) - indic.pct_of(price * nb_coin, fees)
    nb_coin = 0
    return nb_coin, money
```

```python
import websocket
try:
    import thread
except ImportError:
    import _thread as thread
import time


def on_message(ws, message):
    print(message)

def on_error(ws, error):
    print(error)

def on_close(ws):
    print("### closed ###")

def on_open(ws):
    def run(*args):
        for i in range(3):
            time.sleep(1)
            ws.send("Hello %d" % i)
        time.sleep(1)
        ws.close()
        print("thread terminating...")
    thread.start_new_thread(run, ())

ws = websocket.WebSocketApp("wss://stream.binance.com:9443/ws/btcusdt@kline_1m",
                              on_open = on_open,
                              on_message = on_message,
                              on_error = on_error,
                              on_close = on_close)

ws.run_forever()
```
